```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Abrir la imagen original
foto = Image.open('bullterrier.jpg')

# Invertir los colores (negativo)
datos = list(foto.getdata())
datos_invertidos = [(255 - datos[x][0], 255 - datos[x][1], 255 - datos[x][2]) for x in range(len(datos))]
imagen_invertida = Image.new('RGB', foto.size)
imagen_invertida.putdata(datos_invertidos)
plt.subplot(1, 2, 1)
plt.imshow(imagen_invertida)

# Convertir la imagen invertida a escala de grises
imagen_invertida_gris = imagen_invertida.convert('L')

# Convertir la imagen en escala de grises invertida a un arreglo numpy
imagen_invertida_gris_array = np.array(imagen_invertida_gris)

# Mostrar la imagen en escala de grises invertida con matplotlib
plt.subplot(1, 2, 2)
plt.imshow(imagen_invertida_gris_array, cmap='gray')

# Guardar la imagen invertida en escala de grises
imagen_invertida_gris_pil = Image.fromarray(imagen_invertida_gris_array)
imagen_invertida_gris_pil.save('bullterrier_invertido_gris.jpg')
```

```python
# Aplicaci칩n interactiva de visualizaci칩n de canales de color
import cv2
import numpy as np

# Cargar la imagen
image = cv2.imread('bullterrier.jpg')  

# Crear ventanas para mostrar los canales
def show_channel(image, channels):
    """Retorna una imagen mostrando solo los canales seleccionados."""
    channel_image = np.zeros_like(image)
    for i, channel in enumerate(channels):
        if channel:
            channel_image[:, :, i] = image[:, :, i]
    return channel_image

# Configuraci칩n inicial
channels = [True, True, True]  # [Rojo, Verde, Azul]

# Mostrar la imagen en una ventana
cv2.imshow('Image Viewer', show_channel(image, channels))

print("Presiona 'R' para alternar el canal Rojo, 'G' para el Verde, 'B' para el Azul")
print("Presiona 'ESC' para salir.")

while True:
    key = cv2.waitKey(0) & 0xFF  # Espera por la entrada del usuario

    # Teclas de control
    if key == 27:  # ESC para salir
        break
    elif key == ord('r'):  # Alternar el canal rojo
        channels[2] = not channels[2]
    elif key == ord('g'):  # Alternar el canal verde
        channels[1] = not channels[1]
    elif key == ord('b'):  # Alternar el canal azul
        channels[0] = not channels[0]

    # Actualizar la imagen mostrada
    cv2.imshow('Image Viewer', show_channel(image, channels))

cv2.destroyAllWindows()
```

```python
#Dibujo de figuras y texto en im치genes

import cv2

# Cargar la imagen
image = cv2.imread('bullterrier.jpg')  # Aseg칰rate de tener el archivo correcto
output_image = image.copy()  # Copia de la imagen original

# Coordenadas del c칤rculo (ejemplo: cara del perro)
center_coordinates = (215, 120)  # Cambia a las coordenadas de la cara en tu imagen
radius = 80
circle_color = (0, 255, 0)  # Verde
thickness = 3

# Dibujar el c칤rculo
cv2.circle(output_image, center_coordinates, radius, circle_color, thickness)

# A침adir el texto
text = "Perro"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
text_color = (255, 0, 0)  # Azul
text_thickness = 2
text_position = (center_coordinates[0] - 50, center_coordinates[1] - radius - 20)
cv2.putText(output_image, text, text_position, font, font_scale, text_color, text_thickness)

# Guardar la imagen resultante
cv2.imwrite('bullterrier_labeled.jpg', output_image)

# Mostrar la imagen modificada
cv2.imshow('Imagen con Circulo y Texto', output_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Aplicaci칩n de umbral binario
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
image = cv2.imread('bullterrier.jpg')  # Cambia el nombre del archivo de tu imagen

# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplicar el umbral binario
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Mostrar la imagen original y la binarizada con matplotlib
# Convertir de BGR (OpenCV) a RGB (matplotlib) para visualizaci칩n correcta
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Crear un gr치fico con dos subgr치ficos (uno para la imagen original y otro para la binarizada)
plt.figure(figsize=(10, 5))

# Mostrar la imagen original
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Imagen Original')
plt.axis('off')  # Ocultar los ejes

# Mostrar la imagen binarizada
plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')  # Mostrar la imagen en escala de grises
plt.title('Imagen Binarizada')
plt.axis('off')  # Ocultar los ejes

# Mostrar el gr치fico
plt.tight_layout()
plt.show()

```

```python
# Programa de dibujo interactivo con eventos de mouse y teclado
import cv2
import numpy as np

# Variables globales
drawing = False  # Para controlar si estamos dibujando
shape = 'circle'  # Forma predeterminada (c칤rculo)
undo_stack = []  # Para almacenar el historial de im치genes para deshacer
img = np.ones((500, 500, 3), dtype=np.uint8) * 255  # Imagen en blanco

# Funciones de evento de OpenCV
def draw_shape(event, x, y, flags, param):
    global drawing, start_x, start_y, img, shape, undo_stack

    if event == cv2.EVENT_LBUTTONDOWN:  # Empezar a dibujar
        drawing = True
        start_x, start_y = x, y
        undo_stack.append(img.copy())  # Guardamos el estado para deshacer

    elif event == cv2.EVENT_MOUSEMOVE:  # Dibujar mientras movemos el mouse
        if drawing:
            img = undo_stack[-1].copy()  # Restaurar 칰ltima imagen antes de dibujar
            if shape == 'circle':
                cv2.circle(img, (start_x, start_y), int(((x - start_x) ** 2 + (y - start_y) ** 2) ** 0.5), (0, 0, 255), -1)
            elif shape == 'rectangle':
                cv2.rectangle(img, (start_x, start_y), (x, y), (0, 255, 0), -1)
            elif shape == 'line':
                cv2.line(img, (start_x, start_y), (x, y), (255, 0, 0), 2)

    elif event == cv2.EVENT_LBUTTONUP:  # Dejar de dibujar
        drawing = False
        if shape == 'circle':
            cv2.circle(img, (start_x, start_y), int(((x - start_x) ** 2 + (y - start_y) ** 2) ** 0.5), (0, 0, 255), -1)
        elif shape == 'rectangle':
            cv2.rectangle(img, (start_x, start_y), (x, y), (0, 255, 0), -1)
        elif shape == 'line':
            cv2.line(img, (start_x, start_y), (x, y), (255, 0, 0), 2)

# Crear ventana
cv2.namedWindow("Drawing App")
cv2.setMouseCallback("Drawing App", draw_shape)

while True:
    # Mostrar la imagen
    cv2.imshow("Drawing App", img)

    # Esperar tecla presionada
    key = cv2.waitKey(1) & 0xFF

    if key == ord('q'):  # Salir del programa
        break
    elif key == ord('r'):  # Cambiar a rect치ngulo
        shape = 'rectangle'
    elif key == ord('c'):  # Cambiar a c칤rculo
        shape = 'circle'
    elif key == ord('l'):  # Cambiar a l칤nea
        shape = 'line'
    elif key == ord('u'):  # Deshacer 칰ltimo cambio
        if len(undo_stack) > 0:
            img = undo_stack.pop()  # Recuperar el estado anterior

    elif key == ord('s'):  # Guardar imagen
        cv2.imwrite('drawing.png', img)
        print("Imagen guardada como 'drawing.png'")

# Cerrar ventana
cv2.destroyAllWindows()
```
